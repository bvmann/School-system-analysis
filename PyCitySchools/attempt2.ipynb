```python
import pandas as pd 
```

```python
school = 'schools_complete.csv'
student = 'students_complete.csv'

school = pd.read_csv(school)
student = pd.read_csv(student)

todoschool = pd.merge(student,school, how='left', on=['school_name','school_name'])

todoschool.head()

```

```python
total_schools = todoschool['school_name'].nunique()

total_students = todoschool['student_name'].count()

total_budget = school['budget'].sum()

average_math = todoschool['math_score'].sum()/todoschool['math_score'].count()
average_read = todoschool['reading_score'].mean()

pass_mathlist = todoschool.loc[(todoschool['math_score']>=70)]
avg_mathpass = (pass_mathlist['math_score'].count()/todoschool['math_score'].count())*100

pass_readlist = todoschool.loc[(todoschool['reading_score'])>=70]
avg_readpass = (pass_readlist['reading_score'].count()/todoschool['reading_score'].count())*100

overall = pass_mathlist.loc[(pass_mathlist['reading_score']>=70)]

overallavg = (overall['student_name'].count()/total_students)*100

districtsummary = pd.DataFrame([[total_schools,total_students,total_budget,average_math,average_read,
                               avg_mathpass,avg_readpass,overallavg]],columns=['Total Schools','Total Students','Total Budget','Average Math Score','Average Reading Score',
                               '% Passing Math','% Passing Reading','% Overall Passing'])

districtsummary

```

```python
school_name = ['Bailey High School','Cabrera High School','Figueroa High School', 'Ford High School', 'Griffin High School', 
              'Hernandez High School','Holden High School', 'Huang High School', 'Johnson High School', 'Pena High School',
              'Rodriguez High School','Shelton High School','Thomas High School','Wilson High School','Wright High School']

schoolgroup = todoschool.groupby(['school_name'])

typeschool=schoolgroup['type'].min()

s_type = typeschool

totalstudents = schoolgroup['Student ID'].count()
s_totalstu = totalstudents.to_list()

totalschoolbudget = schoolgroup['budget'].min()
s_totsbudg= totalschoolbudget.to_list()

perstudentbudget = (totalschoolbudget/totalstudents)
s_perstubud = perstudentbudget.to_list()

avgmath = schoolgroup['math_score'].mean()
s_avgmath = avgmath.to_list()

avgread= schoolgroup['reading_score'].mean()
s_avgread = avgread.to_list()

passmbyschool = pass_mathlist.groupby(['school_name'])

avgpass_mathscr_school = passmbyschool['math_score'].count()/schoolgroup['Student ID'].count() * 100
s_mathpassavg=avgpass_mathscr_school.to_list()

passrbyschool = pass_readlist.groupby(['school_name'])

avgpass_readscr_school = passrbyschool['reading_score'].count()/schoolgroup['Student ID'].count()*100
s_readpassavg = avgpass_readscr_school.to_list()

overall_group = overall.groupby(['school_name']).count()
    
overall_passing = overall_group['reading_score']/totalstudents*100  

school_summary = pd.DataFrame({'School Name' : school_name,
                              'School Type' : s_type,
                             'Total Students' : totalstudents,
                              'Total School Budget' : totalschoolbudget,
                              'Per Student Budget' : perstudentbudget,
                              'Average Math Score' : avgmath,
                              'Average Reading Score' : avgread,
                              '% Passing Math' : avgpass_mathscr_school,
                              '% Passing Reading' : avgpass_readscr_school,
                              '% Overall Passing' :overall_passing })

school_summary.set_index('School Name')
```
